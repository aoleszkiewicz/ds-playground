```python
import numpy as np
```

```python
data = np.linspace(-1, 1, 50)
```

```python
def mean(data: np.ndarray) -> np.float32:
    return np.sum(data) / len(data)
```

```python
print(mean(data), np.mean(data))
```

```python
def median(data: np.ndarray):
    n = len(data)
    if n % 2 != 0:
        index = int(n / 2)
        return data[index]
    else:
        index_1 = int((n / 2) - 1)
        index_2 = int(n / 2)
        return (data[index_1] + data[index_2]) / 2
```

```python
print(median(data), np.median(data))
```
